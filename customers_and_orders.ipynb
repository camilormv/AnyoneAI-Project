```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Rutas de los archivos CSV
csv_path_customers = r'C:\Users\emava\Documents\PLASS\AnyoneAI\customers_updated.csv'
csv_path_orders = r'C:\Users\emava\Documents\PLASS\AnyoneAI\orders.csv'

# Leer los archivos CSV y crear los datasets
dataset_customers = pd.read_csv(csv_path_customers)
dataset_orders = pd.read_csv(csv_path_orders)

# Convertir la columna 'Date' a tipo datetime
dataset_orders['Date'] = pd.to_datetime(dataset_orders['Date'])

# Filtrar los pedidos de 2021
orders_2021 = dataset_orders[dataset_orders['Date'].dt.year == 2021]

# Agrupar por CustomerID y sumar el OrderTotal
customer_spending_2021 = orders_2021.groupby('CustomerID')['OrderTotal'].sum().reset_index()

# Combinar con el dataset de clientes para obtener los nombres
result = pd.merge(customer_spending_2021, dataset_customers[['CustomerID', 'FirstName', 'LastName']], on='CustomerID', how='left')

# Crear una nueva columna con el nombre completo
result['FullName'] = result['FirstName'] + ' ' + result['LastName']

# Ordenar los clientes por el total gastado en orden descendente
result_sorted = result.sort_values('OrderTotal', ascending=False)

# Clientes a resaltar
clients_to_highlight = ['Brandon Divas', 'Raul Herrera', 'Sophie Labonte', 'Frances Marquez', 'Kenneth Jones']

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(15, 10))

# Colores para las barras
colors = ['#1f77b4' if full_name not in clients_to_highlight else '#d62728' for full_name in result_sorted['FullName']]

# Dibujar las barras
bars = ax.bar(result_sorted['FullName'], result_sorted['OrderTotal'], color=colors)

# Añadir etiquetas y título
ax.set_xlabel('Clientes')
ax.set_ylabel('Total de Compras en 2021')
ax.set_title('Total de Compras del 2021 por Cliente')

# Rotar las etiquetas del eje x para mejor legibilidad
plt.xticks(rotation=90)

# Añadir flechas más grandes y nombres para los clientes resaltados
for i, bar in enumerate(bars):
    if result_sorted['FullName'].iloc[i] in clients_to_highlight:
        ax.annotate(result_sorted['FullName'].iloc[i], xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                    xytext=(0, 20), textcoords='offset points',
                    arrowprops=dict(arrowstyle='->', lw=2.0, mutation_scale=20),
                    ha='center', va='bottom')

# Añadir leyenda
highlight_patch = mpatches.Patch(color='#d62728', label='Clientes Resaltados')
normal_patch = mpatches.Patch(color='#1f77b4', label='Otros Clientes')
ax.legend(handles=[highlight_patch, normal_patch])

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Ruta del archivo CSV de órdenes
csv_path_orders = r'C:\Users\emava\Documents\PLASS\AnyoneAI\orders.csv'

# Leer el archivo CSV de órdenes y crear el dataset
dataset_orders = pd.read_csv(csv_path_orders)

# Convertir la columna 'Date' a tipo datetime
dataset_orders['Date'] = pd.to_datetime(dataset_orders['Date'])

# Agrupar por mes y año, y sumar el OrderTotal
monthly_sales = dataset_orders.groupby([dataset_orders['Date'].dt.year.rename('Year'), dataset_orders['Date'].dt.month.rename('Month')])['OrderTotal'].sum().reset_index()

# Renombrar las columnas para mayor claridad (esto ya no es necesario ya que se hizo en el paso anterior)
# monthly_sales.columns = ['Year', 'Month', 'TotalSales']

# Agrupar por mes y calcular el promedio de ventas a lo largo de los años
average_monthly_sales = monthly_sales.groupby('Month')['OrderTotal'].mean().reset_index()

# Encontrar el mes con las ventas promedio más altas
best_month = average_monthly_sales.loc[average_monthly_sales['OrderTotal'].idxmax()]

# Convertir el número del mes a nombre del mes
import calendar
best_month_name = calendar.month_name[int(best_month['Month'])]

# Imprimir el resultado
print(f"El mejor mes para las ventas históricamente es {best_month_name} con un promedio de ventas de ${best_month['OrderTotal']:.2f}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar

# Ruta del archivo CSV de órdenes
csv_path_orders = r'C:\Users\emava\Documents\PLASS\AnyoneAI\orders.csv'

# Leer el archivo CSV de órdenes y crear el dataset
dataset_orders = pd.read_csv(csv_path_orders)

# Convertir la columna 'Date' a tipo datetime
dataset_orders['Date'] = pd.to_datetime(dataset_orders['Date'])

# Agrupar por mes y año, y sumar el OrderTotal
monthly_sales = dataset_orders.groupby([dataset_orders['Date'].dt.year.rename('Year'), dataset_orders['Date'].dt.month.rename('Month')])['OrderTotal'].sum().reset_index()

# Calcular el promedio de ventas por mes a lo largo de todos los años
average_monthly_sales = monthly_sales.groupby('Month')['OrderTotal'].mean().reset_index()

# Convertir el número del mes a nombre del mes
average_monthly_sales['MonthName'] = average_monthly_sales['Month'].apply(lambda x: calendar.month_abbr[int(x)])

# Crear el gráfico
plt.figure(figsize=(15, 8))
plt.plot(average_monthly_sales['MonthName'], average_monthly_sales['OrderTotal'], marker='o')

# Añadir etiquetas y título
plt.xlabel('Mes')
plt.ylabel('Promedio de Ventas')
plt.title('Promedio de Ventas Históricas por Mes')

# Rotar las etiquetas del eje x para mejor legibilidad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
